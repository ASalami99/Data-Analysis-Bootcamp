```python
import pandas as pd
import matplotlib.pyplot as plt
#Seaborn is another library used for visualisation
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv('bar_chart_data.csv')
```

```python
df_used_cars
```

```python
#Shift + tab to get more info
#plt.figure(figsize=(9,6)) is to set the graph to specific size 9,6. It should usually be enough for most graphs
#plt.xticks(rotation = 45) - makes the labels at a 45 degree angle to manage space better
#you can give different bars different colours by: color = "rgb", where r for red etc

plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"],
       height= df_used_cars["Cars Listings"],
       color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Cars Listing By Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png")
```
