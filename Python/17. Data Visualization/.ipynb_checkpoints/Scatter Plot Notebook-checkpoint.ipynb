```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("scatter_data.csv")
```

```python
df_real_estate
```

```python
#To add a legend, we assigned the graph to a variable 'scatter'

plt.figure(figsize= (12,8))
scatter = plt.scatter(df_real_estate["Area (ft.)"],
            df_real_estate["Price"],
            alpha = 0.6, #Used for transparency
            c = df_real_estate["Building Type"],   #Used for colour coding
            cmap = "viridis")  #To use a different color code from the default
plt.legend(*scatter.legend_elements(),
           loc = "upper left",
           title = "Building Type")
plt.title("Relationship between Area and Price for California Real Estate", fontsize = 14, fontweight = "bold")
plt.xlabel("Area (sq. ft)", fontweight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

## Using Seaborn Library

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(data = df_real_estate,
                x = "Price",
                y = "Area (ft.)",
                hue = "Building Type",   #'Hue' is used for color coding (coloring based on a criteria e.g Building Type) in seaborn.
                palette = ["black", "darkblue", "purple", "pink", "white"],
                s = 100) #The size of each dot
plt.title("Relationship between Area and Price for California Real Estate", fontsize = 14, fontweight = "bold")
plt.xlabel("Area (sq. ft)", fontweight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```
