```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
```

```python
df_ad_expenditure
```

```python
#plt.figure(figsize=(10, 8))
sns.set(rc = {'figure.figsize': (9,6)})  #Another way to change size
sns.regplot(x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            #color = "grey")   #Changes the color of EVERYTHING
            scatter_kws= {'color': 'k'},   #This is for color of the scattar (dots)
            line_kws= {'color': 'red'})    #Color for the line
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
sns.lmplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure,
           height= 9,
           scatter_kws= {'color': 'k'},   #This is for color of the scattar (dots)
           line_kws= {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 16, fontweight = "bold")
plt.show()
```
