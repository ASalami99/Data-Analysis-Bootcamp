```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_stock_prices = pd.read_csv("line_chart_data.csv")
```

```python
df_stock_prices
```

```python
df_stock_prices['new_Date'] = pd.to_datetime(df_stock_prices['Date'])
```

```python
df_stock_prices['new_Date']
```

```python
#X-axis first, then y-axis
labels = ["S&P 500", "FTSE100"]
plt.figure(figsize = (20, 8))
plt.plot(df_stock_prices['new_Date'], df_stock_prices['GSPC500'])
plt.plot(df_stock_prices['new_Date'], df_stock_prices['FTSE100'])
plt.title("S&P Vs FTSE Returns (2000 - 2010)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
#Selecting values for Q3 and Q4 of 2008, basically zooming in on a particular part to gain some insights. The way it is originally, one can barely gain insights.
df_stock_prices_Q2 = df_stock_prices[(df_stock_prices['new_Date'] >= '2008-07-01') &
                                     (df_stock_prices['new_Date'] <= '2008-12-31')]
```

```python
df_stock_prices_Q2
```

```python
#X-axis first, then y-axis
#labels = ["S&P 500", "FTSE100"] <- No need to redefine it because we already did so earlier
plt.figure(figsize = (20, 8))
plt.plot(df_stock_prices_Q2['new_Date'], df_stock_prices_Q2['GSPC500'], color = "midnightblue")
plt.plot(df_stock_prices_Q2['new_Date'], df_stock_prices_Q2['FTSE100'], color = "crimson")
plt.title("S&P Vs FTSE Returns (H2 2008)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
