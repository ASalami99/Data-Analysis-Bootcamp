```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
```

```python
#'Python Users' is a percentage
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize = (10, 7))

#For the bar chart
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "k")
ax.set_ylabel("Number Of Participants", fontweight = "bold")
ax.tick_params(axis = "y",
               width = 2,
               labelsize = "large")

#For the line chart
ax1 = ax.twinx()
ax1.set_ylim(0, 1)   #This and the one below it to convert the percentages (labels on the right)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"],
         color = "#B60000",
         marker = "D")  #Shape of the dots
ax1.set_ylabel("Python Users", color = "#B60000", fontweight = "bold")
ax.tick_params(axis = "y",
               width = 2,
               color = "#B60000",
               labelsize = "large")
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = 14, fontweight = "bold")
```
