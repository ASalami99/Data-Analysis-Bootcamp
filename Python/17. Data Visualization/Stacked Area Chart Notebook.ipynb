```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
```

```python
df_fuel_engine_types
```

```python
#The first thing written (year) is what is on the x-axis. Then you stack on top of that
colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["Diesel", "Petrol", "Gas"] #Trying to add a legend
sns.set_style("white") #Setting a white background

plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor = 'none') #To remove the the outline
plt.xticks(df_fuel_engine_types["Year"], rotation = 45) #To make sure ALL the years are displayed
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number Of Cars", fontsize = 13)
plt.title("Popularity Of Engine Fuel Types (1982 - 2016)", fontsize = 14, fontweight = "bold")
sns.despine() #Removes top and right borders
plt.show()
```
